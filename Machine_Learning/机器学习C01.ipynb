# 绪论


## 基本概念

- 机器学习是研究关于“学习算法”的学问
- 模型泛指从数据中学得的结果
- 数据记录的集合被成做一个样本。属性张成的空间被称为属性空间。实例可以投影在特征空间，因此也可以叫做特征向量。
- 属性的种类数量是样本的维数

从样本中学得模型的过程被称为学习或者训练。训练过程中使用的数据被称为**训练数据**，其中每一个样本叫做训练样本。训练样本组成的集合成为**训练集**。

对模型对应了关于数据的某种潜在的规律的推测，被称为**假设**。潜在规律本身被称为**真相**。

- 如果输出空间是离散的，则属于分类问题
- 如果输出空间是连续的，则属于回归问题

学得模型后，使用其进行预测的过程被称为测试。被测试的样本被称为测试样本。

此外机器学习任务还有聚类。在聚类学习中，训练样本不包括标记信息。

机器学习的目标是使得学得的模型能很好地适用于新样本。能适用于整个样本空间的模型被认为泛化能力强。

通常假设样本空间中全体样本服从一个**未知分布**，获得的每个样本都是独立地从这个分布上采样获得，即是**独立同分布**。一般来说，训练样本越多，得到的关于$D$的信息越多，越有可能通过学习获得具有强泛化能力的模型。

## 假设空间
机器学习是从样例中学习，属于归纳学习。

可以把学习过程看作一个在所有假设组成的空间中进行搜索的过程。搜索目标是找到与训练集匹配的假设。假设的**表示**一旦确定，**假设空间**的规模大小就确定了。注意：可能有多个假设与训练集一致，也就是说，存在者一个与训练集一致的假设集合（**版本空间**）。任何一个有效的机器学习算法必然有其归纳偏好，否则将被假设空间中看似在训练集上等效的假设所迷惑，而无法产生确定的学习结果。

奥卡姆剃刀：若有多个假设与观察一致，选最简单的那个。
